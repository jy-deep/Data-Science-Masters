# 1: What is Matplotlib? Why is it used? Name five plots that can be plotted using the Pyplot module of Matplotlib.
ANSWER:

Matplotlib is a Python library used for creating static, animated, and interactive visualizations in Python. It provides a wide range of options for creating visualizations and can be used in various scientific, engineering, and data analysis applications. Matplotlib is open-source and can be used freely.

Matplotlib is used for creating different types of plots, including line plots, scatter plots, bar plots, histogram plots, and pie charts. It allows users to customize plots with various formatting options, including colors, labels, markers, and fonts.

The Pyplot module of Matplotlib is used for creating simple plots and visualizations quickly. Some of the plots that can be plotted using Pyplot module of Matplotlib are:

    Line plot: A line plot is used to visualize the relationship between two variables. It is created by plotting a series of data points on a line.

    Scatter plot: A scatter plot is used to visualize the relationship between two variables. It is created by plotting individual data points on a two-dimensional plane.

    Bar plot: A bar plot is used to visualize the distribution of categorical data. It is created by plotting bars of different heights that represent the frequency or proportion of each category.

    Histogram plot: A histogram plot is used to visualize the distribution of continuous data. It is created by dividing the range of data into equal intervals and plotting the frequency or proportion of data points in each interval.

    Pie chart: A pie chart is used to visualize the relative proportion of different categories in a dataset. It is created by dividing a circle into slices that represent the proportion of each category.

# 2: What is a scatter plot? Use the following code to generate data for x and y. Using this generated data plot a scatter plot.
    import numpy as np
    np.random.seed(3)
    x = 3 + np.random.normal(0, 2, 50)
    y = 3 + np.random.normal(0, 2, len(x))
ANSWER:

A scatter plot is a type of plot used to display the relationship between two continuous variables. It consists of a set of data points, where each point represents the values of the two variables. The position of each point on the plot corresponds to the values of the two variables for that point, with one variable represented on the x-axis and the other variable represented on the y-axis.

Here's the code to generate a scatter plot for the given data:

    import numpy as np
    import matplotlib.pyplot as plt

    np.random.seed(3)

    x = 3 + np.random.normal(0, 2, 50)
    y = 3 + np.random.normal(0, 2, len(x))

    plt.scatter(x, y)
    plt.title("Scatter Plot")
    plt.xlabel("X values")
    plt.ylabel("Y values")
    plt.show()


# 3: Why is the subplot() function used? Draw four line plots using the subplot() function.
    Use the following data:
    import numpy as np
    For line 1: x = np.array([0, 1, 2, 3, 4, 5]) and y = np.array([0, 100, 200, 300, 400, 500])
    For line 2: x = np.array([0, 1, 2, 3, 4, 5]) and y = np.array([50, 20, 40, 20, 60, 70])
    For line 3: x = np.array([0, 1, 2, 3, 4, 5]) and y = np.array([10, 20, 30, 40, 50, 60])
    For line 4: x = np.array([0, 1, 2, 3, 4, 5]) and y = np.array([200, 350, 250, 550, 450, 150])
ANSWER:

The subplot() function in Matplotlib is used to create multiple subplots in a single figure. It allows us to display multiple plots side by side, making it easier to compare and analyze data.

The subplot() function takes three arguments: the number of rows, the number of columns, and the index of the subplot to create. The index starts from 1 and goes from left to right, top to bottom.

Here's the code to create four line plots using the subplot() function:

    import numpy as np
    import matplotlib.pyplot as plt

    # Data for line 1
    x1 = np.array([0, 1, 2, 3, 4, 5])
    y1 = np.array([0, 100, 200, 300, 400, 500])

    # Data for line 2
    x2 = np.array([0, 1, 2, 3, 4, 5])
    y2 = np.array([50, 20, 40, 20, 60, 70])

    # Data for line 3
    x3 = np.array([0, 1, 2, 3, 4, 5])
    y3 = np.array([10, 20, 30, 40, 50, 60])

    # Data for line 4
    x4 = np.array([0, 1, 2, 3, 4, 5])
    y4 = np.array([200, 350, 250, 550, 450, 150])

    # Create a 2x2 grid of subplots
    fig, axs = plt.subplots(2, 2)

    # Plot line 1
    axs[0, 0].plot(x1, y1)
    axs[0, 0].set_title('Line 1')

    # Plot line 2
    axs[0, 1].plot(x2, y2)
    axs[0, 1].set_title('Line 2')

    # Plot line 3
    axs[1, 0].plot(x3, y3)
    axs[1, 0].set_title('Line 3')

    # Plot line 4
    axs[1, 1].plot(x4, y4)
    axs[1, 1].set_title('Line 4')

    # Add space between subplots
    plt.subplots_adjust(wspace=0.3, hspace=0.3)

    # Display the plots
    plt.show()


# 4: What is a bar plot? Why is it used? Using the following data plot a bar plot and a horizontal bar plot.
    import numpy as np
    company = np.array(["Apple", "Microsoft", "Google", "AMD"])
    profit = np.array([3000, 8000, 1000, 10000])
ANSWER:

A bar plot is a type of plot used to display categorical data with rectangular bars. The length of each bar is proportional to the value being represented. Bar plots are commonly used to compare data across different categories or to show the distribution of a categorical variable.

The bar() function in Matplotlib can be used to create vertical bar plots, and the barh() function can be used to create horizontal bar plots.

Here's the code to create a vertical bar plot and a horizontal bar plot using the given data:

    import numpy as np
    import matplotlib.pyplot as plt

    # Data
    company = np.array(["Apple", "Microsoft", "Google", "AMD"])
    profit = np.array([3000, 8000, 1000, 10000])

    # Vertical bar plot
    plt.bar(company, profit)
    plt.title('Company profits')
    plt.xlabel('Company')
    plt.ylabel('Profit (in millions)')
    plt.show()

    # Horizontal bar plot
    plt.barh(company, profit)
    plt.title('Company profits')
    plt.xlabel('Profit (in millions)')
    plt.ylabel('Company')
    plt.show()


# 5: What is a box plot? Why is it used? Using the following data plot a box plot.
    box1 = np.random.normal(100, 10, 200)
    box2 = np.random.normal(90, 20, 200)
ANSWER:

A box plot is a type of plot that displays the distribution of a dataset. It is also known as a box-and-whisker plot because it consists of a box that represents the middle 50% of the data and "whiskers" that represent the variability outside the box. Box plots are useful for identifying outliers, comparing distributions, and detecting skewness or symmetry in the data.

To create a box plot in Matplotlib, we can use the boxplot() function. Here's the code to create a box plot for the given data:

    import numpy as np
    import matplotlib.pyplot as plt

    # Data
    box1 = np.random.normal(100, 10, 200)
    box2 = np.random.normal(90, 20, 200)

    # Create a figure and axis
    fig, ax = plt.subplots()

    # Create the box plot
    ax.boxplot([box1, box2])

    # Add title and axis labels
    ax.set_title('Box Plot')
    ax.set_xlabel('Box')
    ax.set_ylabel('Value')

    # Show the plot
    plt.show()
