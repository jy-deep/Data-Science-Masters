# 2. What is the difference between the functions loc and iloc in pandas.DataFrame?
ANSWER:

In Pandas, loc and iloc are both methods used to access data in a DataFrame, but they differ in how they handle indexing and slicing.

    loc: It is used for label-based indexing, where you specify the label (or the index name) to access the data. It can be used with either row or column labels, or a combination of both. The syntax for using loc is df.loc[row_label, column_label], where row_label can be a single label, a list of labels, or a slice of labels, and column_label can be a single label, a list of labels, or a slice of labels.

Example:

    # Access data in the row with label '1' and column with label 'duration'
    value = df.loc[1, 'duration']
    print(value)
    # Output: 3

    iloc: It is used for integer-based indexing, where you specify the integer index to access the data. It can be used with either integer row or column indices, or a combination of both. The syntax for using iloc is df.iloc[row_index, column_index], where row_index can be a single integer, a list of integers, or a slice of integers, and column_index can be a single integer, a list of integers, or a slice of integers.

Example:

    # Access data in the second row (index 1) and first column (index 0)
    value = df.iloc[1, 0]
    print(value)
    # Output: 'Machine Learning'

# 6. What do you understand about the windows function in pandas and list the types of windows functions?
ANSWER:

In pandas, the window functions are used for performing calculations on a sliding or expanding window of data within a DataFrame or Series. These functions allow us to compute various statistical measures, such as mean, median, sum, etc., on a rolling or expanding window of data.

There are several types of window functions in pandas, including:

    Rolling Window Functions: These functions operate on a fixed-size sliding window of data and provide calculations based on that window. Some commonly used rolling window functions are:
        rolling.mean(): calculates the mean of the values in the window
        rolling.sum(): calculates the sum of the values in the window
        rolling.max(): calculates the maximum value in the window
        rolling.min(): calculates the minimum value in the window
        rolling.std(): calculates the standard deviation of the values in the window

    Expanding Window Functions: These functions operate on an expanding window of data, where the size of the window increases with each data point. Some commonly used expanding window functions are:
        expanding.mean(): calculates the mean of all the values up to the current data point
        expanding.sum(): calculates the sum of all the values up to the current data point
        expanding.max(): calculates the maximum value of all the values up to the current data point
        expanding.min(): calculates the minimum value of all the values up to the current data point
        expanding.std(): calculates the standard deviation of all the values up to the current data point

# 8. Write a Python program that takes in two dates as input (in the format YYYY-MM-DD) and calculates the difference between them in days, hours, and minutes using Pandas time delta. The program should prompt the user to enter the dates and display the result.
ANSWER:

    import pandas as pd

    # Get input dates from user
    date1 = input("Enter date 1 (YYYY-MM-DD): ")
    date2 = input("Enter date 2 (YYYY-MM-DD): ")

    # Convert input dates to datetime objects
    date1 = pd.to_datetime(date1)
    date2 = pd.to_datetime(date2)

    # Calculate time delta between the two dates
    time_delta = date2 - date1

    # Extract days, hours, and minutes from time delta
    days = time_delta.days
    hours = time_delta.seconds // 3600
    minutes = (time_delta.seconds // 60) % 60

    # Display the result
    print("Difference between the two dates:")
    print("Days: ", days)
    print("Hours: ", hours)
    print("Minutes: ", minutes)


# 9. Write a Python program that reads a CSV file containing categorical data and converts a specified column to a categorical data type. The program should prompt the user to enter the file path, column name, and category order, and then display the sorted data.
ANSWER:

    import pandas as pd

    # Get file path from user
    file_path = input("Enter file path: ")

    # Read CSV file
    df = pd.read_csv(file_path)

    # Get column name from user
    column_name = input("Enter column name to convert to categorical: ")

    # Convert column to categorical data type
    df[column_name] = pd.Categorical(df[column_name])

    # Get category order from user
    category_order = input("Enter category order (comma-separated): ").split(",")

    # Set category order for the column
    df[column_name] = df[column_name].cat.set_categories(category_order)

    # Sort the data by the specified column
    df.sort_values(column_name, inplace=True)

    # Display the sorted data
    print("Sorted data by column", column_name)
    print(df)


# 10. Write a Python program that reads a CSV file containing sales data for different products and visualizes the data using a stacked bar chart to show the sales of each product category over time. The program should prompt the user to enter the file path and display the chart.
ANSWER:

    import pandas as pd
    import matplotlib.pyplot as plt

    # Get file path from user
    file_path = input("Enter file path: ")

    # Read CSV file
    df = pd.read_csv(file_path)

    # Group data by product category and time period
    grouped_data = df.groupby(['product_category', 'time_period'])['sales'].sum().unstack()

    # Plot stacked bar chart
    ax = grouped_data.plot(kind='bar', stacked=True, figsize=(10, 6))
    ax.set_xlabel('Time Period')
    ax.set_ylabel('Sales')
    ax.set_title('Sales by Product Category over Time')
    plt.legend(title='Product Category')
    plt.show()

ANSWER:

    import pandas as pd

    # Get file path from user
    file_path = input("Enter the file path of the CSV file containing the student data: ")

    # Read CSV file into a Pandas DataFrame
    df = pd.read_csv(file_path)

    # Calculate mean, median, and mode of test scores
    mean_score = df['Test Score'].mean()
    median_score = df['Test Score'].median()
    mode_score = df['Test Score'].mode().tolist()

    # Display the results in a table
    results = pd.DataFrame({'Statistic': ['Mean', 'Median', 'Mode'], 'Value': [mean_score, median_score, mode_score]})
    print(results)
