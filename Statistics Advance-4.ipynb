# 1: What is the difference between a t-test and a z-test? Provide an example scenario where you would use each type of test.
ANSWER:

Both the t-test and the z-test are statistical hypothesis tests used to make inferences about population parameters based on sample data. However, they differ in their assumptions and when they should be used.

1. t-test:
   - The t-test is used when the population standard deviation is unknown and must be estimated from the sample data.
   - It is appropriate when the sample size is relatively small (typically less than 30) or when the population distribution is not normally distributed.
   - The test statistic in a t-test follows a t-distribution.
   - Example scenario: Suppose you want to compare the mean heights of two groups of students, where the standard deviations of the heights are not known. You would use a t-test to determine if the difference in means is statistically significant.

2. z-test:
   - The z-test is used when the population standard deviation is known or when the sample size is large (typically greater than 30).
   - It assumes that the population follows a normal distribution.
   - The test statistic in a z-test follows a standard normal distribution (z-distribution).
   - Example scenario: Consider a situation where you want to test if the mean weight of a sample of individuals is significantly different from a known population mean weight. If the population standard deviation is known, you can use a z-test to compare the means.

# 2: Differentiate between one-tailed and two-tailed tests.
ANSWER:

One-Tailed Test:
In a one-tailed test, the alternative hypothesis is directional, meaning it specifies a specific direction of the effect or difference being tested. The test is designed to determine whether the observed data is significantly greater than or less than a certain value.

Example: Suppose you want to test whether a new drug improves test scores compared to a placebo. The one-tailed hypothesis would state that the mean test score with the drug is greater than the mean test score with the placebo.

Two-Tailed Test:
In a two-tailed test, the alternative hypothesis is non-directional, meaning it does not specify a particular direction of the effect or difference. The test aims to determine whether the observed data is significantly different from a certain value, without indicating if it is greater or smaller.

Example: Imagine you want to test whether a new teaching method affects students' test scores. The two-tailed hypothesis would state that there is a difference in the mean test scores between the two teaching methods, without specifying if one method is expected to be superior.

Key differences:
1. Hypothesis formulation: In a one-tailed test, the alternative hypothesis specifies the direction of the effect, while in a two-tailed test, it does not specify a direction.
2. Critical regions: In a one-tailed test, the critical region is located entirely on one side of the distribution (either the upper or lower tail). In a two-tailed test, the critical region is divided between both tails of the distribution.
3. P-value interpretation: In a one-tailed test, the p-value represents the probability of observing a result as extreme as the one obtained, in the specified direction. In a two-tailed test, the p-value represents the probability of observing a result as extreme as the one obtained, in either direction.

The choice between a one-tailed and a two-tailed test depends on the specific research question, the directionality of the effect, and prior knowledge or hypotheses about the phenomenon under investigation. It is essential to carefully consider the appropriate test to ensure the hypothesis is adequately addressed.

# 3: Explain the concept of Type 1 and Type 2 errors in hypothesis testing. Provide an example scenario for each type of error.
ANSWER:

In hypothesis testing, Type 1 and Type 2 errors are two possible mistakes that can occur when interpreting the results.

Type 1 Error:
A Type 1 error, also known as a false positive, occurs when the null hypothesis is incorrectly rejected, even though it is true. In other words, it is the error of concluding there is a significant effect or difference when, in reality, there is none.

Example scenario:
Suppose a pharmaceutical company is conducting clinical trials to test the effectiveness of a new drug for a specific disease. The null hypothesis states that the drug has no effect on the disease. If the researchers erroneously reject the null hypothesis and conclude that the drug is effective (significant effect), but it is, in fact, ineffective, it would be a Type 1 error. This could lead to the drug being approved and released to the market when it does not provide any real benefit.

Type 2 Error:
A Type 2 error, also known as a false negative, occurs when the null hypothesis is incorrectly accepted, even though it is false. It is the error of failing to detect a significant effect or difference when one truly exists.

Example scenario:
Consider a scenario where a manufacturing company claims that their new manufacturing process reduces the defect rate of their products by 50%. The null hypothesis states that the defect rate remains the same. If the researchers fail to reject the null hypothesis and conclude that there is no significant effect (no difference in defect rates), but the new process does indeed reduce defects, it would be a Type 2 error. This could result in the company missing out on the opportunity to improve their manufacturing process and reduce costs.

It's important to note that Type 1 and Type 2 errors are inversely related. Reducing the likelihood of one type of error typically increases the likelihood of the other. The significance level (alpha) chosen for the test and the sample size influence the probability of committing these errors. Researchers aim to strike a balance between these two types of errors based on the consequences and objectives of the study.

# 4: Explain Bayes's theorem with an example.
ANSWER:

Bayes's theorem is a fundamental concept in probability theory that allows us to update our beliefs or probabilities based on new evidence or information. It provides a way to calculate the probability of an event given prior knowledge and new data.

The theorem is expressed as:

P(A|B) = (P(B|A) * P(A)) / P(B)

Where:
P(A|B) is the posterior probability of event A given event B (the probability we want to calculate).
P(B|A) is the probability of event B given event A (likelihood).
P(A) is the prior probability of event A (our initial belief in the absence of evidence).
P(B) is the probability of event B (normalizing constant).

Example scenario:
Let's consider a medical example to understand Bayes's theorem. Suppose we have a rare disease that affects 1 in 10,000 people. We also know that the diagnostic test for this disease is 99% accurate in both identifying those who have the disease (true positive) and those who don't (true negative).

Let's calculate the probability that a person has the disease given that they tested positive (P(A|B)). We can break it down using Bayes's theorem:

P(A) = 1/10,000 (prior probability of having the disease)
P(B|A) = 0.99 (probability of testing positive given the person has the disease)
P(B) = (P(B|A) * P(A)) + (P(B|~A) * P(~A))
        = (0.99 * 1/10,000) + (0.01 * 9,999/10,000) 
        = 0.000099 + 0.9999 = 1

Now, let's calculate the posterior probability using Bayes's theorem:

P(A|B) = (P(B|A) * P(A)) / P(B)
        = (0.99 * 1/10,000) / 1
        = 0.000099

Therefore, the probability that a person has the disease given that they tested positive is approximately 0.000099 or 0.0099%.

# 5: What is a confidence interval? How to calculate the confidence interval, explain with an example.
ANSWER:

A confidence interval is a range of values constructed from sample data that is likely to contain the true population parameter with a certain level of confidence. It provides a measure of uncertainty associated with estimating an unknown parameter based on a sample.

To calculate a confidence interval, you need the following information:
1. Sample data: This is the data collected from a sample of the population.
2. Confidence level: This represents the desired level of confidence (typically expressed as a percentage) that the interval will contain the true population parameter.

Here's an example to illustrate how to calculate a confidence interval:

Suppose you want to estimate the average height of a certain population. You collect a sample of 100 individuals and measure their heights. Let's assume the sample mean height is 170 cm, and the sample standard deviation is 5 cm.

1. Determine the confidence level: Let's say you choose a 95% confidence level, which is a common choice. This means you want to construct an interval that has a 95% chance of containing the true population mean.

2. Determine the critical value: The critical value corresponds to the chosen confidence level and the distribution of the data. For a normal distribution, like in this case, you can use the Z-score associated with the desired confidence level. For a 95% confidence level, the critical value is approximately 1.96.

3. Calculate the standard error: The standard error represents the standard deviation of the sample mean and is calculated as the sample standard deviation divided by the square root of the sample size. In this example, the standard error is 5 / sqrt(100) = 0.5.

4. Calculate the margin of error: The margin of error is determined by multiplying the critical value by the standard error. In this case, the margin of error is 1.96 * 0.5 = 0.98.

5. Calculate the confidence interval: The confidence interval is calculated by adding and subtracting the margin of error from the sample mean. In this example, the confidence interval is (170 - 0.98, 170 + 0.98) = (169.02, 170.98).

Interpretation: With a 95% confidence level, you can say that you are 95% confident that the true average height of the population falls within the interval of 169.02 cm to 170.98 cm.

The confidence interval provides a range within which the true population parameter is likely to lie, considering the variability in the sample data and the chosen level of confidence. The larger the sample size and the higher the confidence level, the narrower the confidence interval becomes, indicating increased precision in the estimation.
