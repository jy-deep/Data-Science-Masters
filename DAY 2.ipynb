# 2: What are the two tuple methods in python? Give an example of each method. Give a reason why tuples have only two in built methods as compared to lists.
ANSWER:

In Python, tuples are a built-in data type and they come with two default methods:
    1.count(): The count() method takes a single argument, which is the element to be counted. It returns an integer that represents the number of times that element appears in the tuple. For example: 
    my_tuple = (1, 2, 3, 1, 2, 3, 1, 2, 3)
    count_one = my_tuple.count(1)
    print(count_one) # Output: 3
    2.index(): The index() method takes a single argument, which is the element to be searched. It returns the index of the first occurrence of that element in the tuple. If the element is not found, it raises a ValueError exception. For example: 
    my_tuple = (1, 2, 3, 1, 2, 3)
    index_two = my_tuple.index(2)
    print(index_two) # Output: 1

Tuples have only two built-in methods as compared to lists because tuples are immutable, meaning that once they are created, they cannot be modified. Therefore, they do not require as many built-in methods as lists, which are mutable data structures and can be modified dynamically.

# 4: Explain the difference between the union() and update() methods for a set. Give an example of each method.
Answer:

union() method returns a new set that contains all the distinct elements from the original set as well as another set or iterable that is passed as an argument to the union() method. The original set remains unchanged. For example:
    set1 = {1, 2, 3}
    set2 = {3, 4, 5}
    set3 = set1.union(set2)
    print(set1)  # {1, 2, 3}
    print(set2)  # {3, 4, 5}
    print(set3)  # {1, 2, 3, 4, 5}
    
update() method modifies the original set by adding all the distinct elements from another set or iterable that is passed as an argument to the update() method. For example:
    set1 = {1, 2, 3}
    set2 = {3, 4, 5}
    set1.update(set2)
    print(set1)  # {1, 2, 3, 4, 5}
    print(set2)  # {3, 4, 5}


# 5: What is a dictionary? Give an example. Also, state whether a dictionary is ordered or unordered.
Answers:

#A dictionary is a built-in data type in Python used to store data in a key-value format. It is an unordered collection of elements where each element is stored as a key-value pair. The keys are unique, and each key is associated with a single value.

### Here's an example of a dictionary:
### Define a dictionary
my_dict = {'apple': 3, 'banana': 5, 'orange': 2}

### Accessing dictionary elements using keys
print(my_dict['apple'])  # Output: 3
print(my_dict['banana'])  # Output: 5
print(my_dict['orange'])  # Output: 2

### Adding a new element to the dictionary
my_dict['pear'] = 4
print(my_dict)  # Output: {'apple': 3, 'banana': 5, 'orange': 2, 'pear': 4}

### Modifying an existing element in the dictionary
my_dict['banana'] = 6
print(my_dict)  # Output: {'apple': 3, 'banana': 6, 'orange': 2, 'pear': 4}

### Removing an element from the dictionary
del my_dict['orange']
print(my_dict)  # Output: {'apple': 3, 'banana': 6, 'pear': 4}


A dictionary is an unordered collection of key-value pairs in Python. This means that the order of the elements in a dictionary is not fixed, and it can change whenever we add, modify or remove elements from the dictionary.

6: Can we create a nested dictionary If so, please give an example by creating a simple one-leveled nested dictionary.
Answers:

Yes, we can create a nested dictionary in Python. A nested dictionary is a dictionary within a dictionary, where the values can be dictionaries themselves. Here's an example of a simple one-level nested dictionary:

### Define a nested dictionary
my_dict = {'person1': {'name': 'Alice', 'age': 25},
           'person2': {'name': 'Bob', 'age': 30}}

### Accessing nested dictionary elements using keys
print(my_dict['person1']['name'])  # Output: 'Alice'
print(my_dict['person2']['age'])  # Output: 30

### Adding a new element to the nested dictionary
my_dict['person1']['city'] = 'New York'
print(my_dict)  # Output: {'person1': {'name': 'Alice', 'age': 25, 'city': 'New York'},
                #          'person2': {'name': 'Bob', 'age': 30}}

### Modifying an existing element in the nested dictionary
my_dict['person2']['age'] = 35
print(my_dict)  # Output: {'person1': {'name': 'Alice', 'age': 25, 'city': 'New York'},
                #          'person2': {'name': 'Bob', 'age': 35}}

### Removing an element from the nested dictionary
del my_dict['person1']['city']
print(my_dict)  # Output: {'person1': {'name': 'Alice', 'age': 25},
                #          'person2': {'name': 'Bob', 'age': 35}}

8: What are the three view objects in dictionaries? Use the three in-built methods in python to display these three view objects for the given dictionary.
dict1 = {'Sport': 'Cricket' , 'Teams': ['India', 'Australia', 'England', 'South Africa', 'Sri Lanka', 'New Zealand']}
Answer:

In Python, dictionaries have three built-in view objects:

dict.keys() - returns a view object that contains the keys of the dictionary.
dict.values() - returns a view object that contains the values of the dictionary.
dict.items() - returns a view object that contains the key-value pairs of the dictionary as tuples.

Here's the code to display all three view objects for the given dictionary:
dict1 = {'Sport': 'Cricket', 'Teams': ['India', 'Australia', 'England', 'South Africa', 'Sri Lanka', 'New Zealand']}

### Displaying keys view object
print("Keys View Object:", dict1.keys())

### Displaying values view object
print("Values View Object:", dict1.values())

### Displaying items view object
print("Items View Object:", dict1.items())

### Output
Keys View Object: dict_keys(['Sport', 'Teams'])
Values View Object: dict_values(['Cricket', ['India', 'Australia', 'England', 'South Africa', 'Sri Lanka', 'New Zealand']])
Items View Object: dict_items([('Sport', 'Cricket'), ('Teams', ['India', 'Australia', 'England', 'South Africa', 'Sri Lanka', 'New Zealand'])])
