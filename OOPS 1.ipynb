3. What is multiple inheritance? Write a python code to demonstrate multiple inheritance.
Answer:

Multiple inheritance is a feature in object-oriented programming where a class can inherit from more than one base class. In Python, a class can inherit from multiple base classes by specifying them in a comma-separated list inside the parentheses of the class definition.

Here's an example code that demonstrates multiple inheritance in Python:

    class Animal:
        def __init__(self, name):
            self.name = name
    
        def speak(self):
            raise NotImplementedError("Subclass must implement abstract method")

    class Pet:
        def __init__(self, breed):
            self.breed = breed
    
        def play(self):
            print(f"The {self.breed} is playing.")

    class Dog(Animal, Pet):
        def __init__(self, name, breed):
            Animal.__init__(self, name)
            Pet.__init__(self, breed)

        def speak(self):
            return "Woof!"

my_dog = Dog("Buddy", "Labrador")
print(my_dog.name)              # Output: Buddy
print(my_dog.breed)             # Output: Labrador
print(my_dog.speak())           # Output: Woof!
my_dog.play()                  # Output: The Labrador is playing.


4. What are getter and setter in python? Create a class and create a getter and a setter method in this class.
Answer:

In Python, getters and setters are methods that are used to retrieve and modify the values of an object's attributes, respectively. They are commonly used to ensure that attribute values are accessed and modified in a controlled way.

Here's an example code that demonstrates how to create a class with a getter and a setter method:

    class Person:
        def __init__(self, name, age):
            self._name = name
            self._age = age

        def get_name(self):
            return self._name

        def set_name(self, name):
            self._name = name

        def get_age(self):
            return self._age

        def set_age(self, age):
            if age >= 0:
                self._age = age
            else:
                raise ValueError("Age cannot be negative.")

person = Person("John", 25)
print(person.get_name())        # Output: John
person.set_name("Jane")
print(person.get_name())        # Output: Jane
print(person.get_age())         # Output: 25
person.set_age(30)
print(person.get_age())         # Output: 30
person.set_age(-10)             # Output: ValueError: Age cannot be negative.


5. What is method overriding in python? Write a python code to demonstrate method overriding.
Answer:

Method overriding is a feature in object-oriented programming where a subclass provides a different implementation of a method that is already defined in its superclass. The subclass can override the method to provide its own implementation and behavior.

Here's an example code that demonstrates method overriding in Python:

    class Animal:
        def __init__(self, name):
            self.name = name
    
        def speak(self):
            print("This animal doesn't have a sound.")

    class Dog(Animal):
        def __init__(self, name):
            Animal.__init__(self, name)

        def speak(self):
            print("Woof!")

    class Cat(Animal):
        def __init__(self, name):
            Animal.__init__(self, name)

        def speak(self):
            print("Meow!")

my_dog = Dog("Buddy")
my_cat = Cat("Mittens")

my_dog.speak()      # Output: Woof!
my_cat.speak()      # Output: Meow!
